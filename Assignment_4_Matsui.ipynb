# The ocean promximity is not in number so, let's encode or mapping it.ラベルエンコーディング = カテゴリを整数にマッピングする

```python
mapping = {'<1H OCEAN': 1,'INLAND': 2,'NEAR OCEAN': 3,'NEAR BAY':4,'ISLAND':5}
df['ocean_proximity'] = df['ocean_proximity'].map(mapping)
```

```python
X = df.drop(['median_house_value'],axis=1)
y = df['median_house_value']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

> この部分では、線形回帰と決定木回帰の r2 (二乗平均平方根値) を見つけ、トレーニングとテストの r2 値を比較して、今日の「過剰適合」と「不足適合」のクラスと関連付けます。
**In this part find the r2, root mean square value of the linear regression and decision tree regressor and compare the r2 value in the training and testing, and connect with your today class on "overfitting" and "underfitting"

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
rmse   = np.sqrt(mean_squared_error(y_test,y_pred))
print('Training Score: ', reg.score(X_train, y_train))
print('Testing Score: ', reg.score(X_test, y_test))
print('RMSE: ',rmse)
```
